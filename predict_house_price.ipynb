```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
```

```python
path="/Users/jeya/Desktop/ML/Real estate.csv"
data=pd.read_csv(path)
data.head()
```

```python
data_numpy=data.values
data_numpy
```

```python
data_tensor=torch.from_numpy(data_numpy)
print(data_tensor.dtype)
data_tensor=data_tensor.to(torch.float32)
print(data_tensor.dtype)
```

```python
inputs=data_tensor[:,1:-1]
print(inputs.shape)
inputs
```

```python
targets=data_tensor[:,-1]
print(targets.shape)
targets=targets.unsqueeze_(1)
print(targets.shape)
```

```python
from torch.utils.data import TensorDataset
tensor_ds=TensorDataset(inputs,targets)
tensor_ds[:9]
```

```python
from torch.utils.data import DataLoader
batch_size=20
trian_loader=DataLoader(tensor_ds,batch_size,shuffle=True)
```

```python
for xb,yb in trian_loader:
    print(xb)
    print(yb)
    break
```

```python
model=nn.Linear(6,1)
print(model.weight)
print(model.bias)
list(model.parameters())
```

```python
preds=model(inputs)
preds
```

```python
import torch.nn.functional as f
loss_fn=f.mse_loss
loss=loss_fn(preds,targets)
loss.data
```

```python
opt=torch.optim.SGD(model.parameters(),1e-9)
```

```python
def fit(num_epochs):
    for epochs in range(num_epochs):
        for xb,yb in trian_loader:
            preds=model(xb)
            loss=loss_fn(preds,yb)
            loss.backward()
            opt.step()
            opt.zero_grad()
        if (epochs+1)%10==0:
            print("current epoch/number of epochs [{}]/[{}],loss : {}".format(epochs+1,num_epochs,loss))
```

```python
fit(100)
```
